```python
import transformers
from transformers import pipeline
```

```python
classifier = pipeline('text-classification')
```

```python
text='the Pineapple Peeler Pro 5000 took longer to use, wasted more fruit, and created more mess than my trusty old knife. Save your money and sanity, folks. Stick to the tried-and-true methods. This "pro" is only a pro at disappointment.'
```

# Emotion classification

```python
import pandas as pd

outputs=classifier(text)
pd.DataFrame(outputs)
```

# Named Entity Recognition (NER)

```python
ner_tagger = pipeline('ner',aggregation_strategy='simple')
outputs = ner_tagger(text)
pd.DataFrame(outputs)
```

# Question-Response

```python
reader = pipeline("question-answering")
question = "Whatd does the customer want?"
outputs=reader(question=question, context=text)
pd.DataFrame([outputs])
```

# Summarization

```python
summarizer = pipeline("summarization")
outputs = summarizer(text, max_length=30, clean_up_tokenization_spaces =True)
print('-------------',outputs[0]['summary_text'])
```

```python
translator = pipeline("translation_en_to_de", model="Helsinki-NLP/opus-mt-en-de")
outputs = translator(text, clean_up_tokenization_spaces=True, min_length=100)
print('-------',outputs[0]['translation_text'])
```

# Generate text

```python
from transformers import set_seed
set_seed(110)
generator = pipeline("text-generation")
response = "Dear customer, I am sorry to hear that pineapple peeler was not helpful."
prompt = text + "\n\nCustomer service response:\n" + response
outputs = generator(prompt, max_length=200)
print(outputs[0]['generated_text'])

```
